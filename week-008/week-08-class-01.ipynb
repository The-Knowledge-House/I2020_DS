We've seen how to use iterables. We've also seen functions like `map`, `filter`, `range` and `functools.reduce` that created iterators. In order to use those iterators, we've casted them as lists. How can we use iterators like designed?

````python
# Turning an iterable into an iterator

l = [1,2,3,4,5]

iterator_l = iter(l)
````

````python
next(iterator_l)
````

#### Behind the scenes, objects that are iterators have a dunder method `__iter__` and `__next__`.

`__iter__` handles how to make the object iterable  
`__next__` handles how to deliver the 'next' element in the iterator.

#### There's an easier way to create iterators: `generator` functions

````python
def foo():
    print("begin")
    for i in range(3):
        print("before yield", i)
        yield i
        print("after yield", i)
    print("end")
````

````python
f = foo()
f
````

````python
for i in f:
    print(i)
````

````python
f = foo()
f
````

````python
next(f)
````

# What happens when you enter a website into your browser and press enter?

https://medium.com/@maneesha.wijesinghe1/what-happens-when-you-type-an-url-in-the-browser-and-press-enter-bb0aa2449c1a

1. You type maps.google.com into the address bar of your browser.
2. The browser checks the cache for a DNS record to find the corresponding IP address of maps.google.com.
3. If the requested URL is not in the cache, ISP’s DNS server initiates a DNS query to find the IP address of the server that hosts maps.google.com.
4. The browser initiates a TCP connection with the server.
5. The browser sends an HTTP request to the webserver.
6. The server handles the request and sends back a response.
7. The server sends out an HTTP response.
8. The browser displays the HTML content (for HTML responses, which is the most common).

# Parts of a URL




# DNS



Domain Name System is a system that matches domain names (e.g. maps.google.com) with a corresponding IP address.

Every server connected to the web has an IP address. This IP address can be static or dynamic. However in both cases, it's usually not best to use as the preferred choice to access the server. Usually you'll configure your DNS name servers through your domain registrars' and web hosts' website. This is done in two parts. The first part is to configure your registrar's information to associate your web domain with a specific server (IP address) and then you'll configure your web hosts' DNS info. The first part is to establish on the domain end that your site is to be linked to a specific IP address and the second part is to allow the domain redirecting to access your server. This process requires propagation of the new DNS info throughout the internet which means changes to DNS information can take up to 24 hours to affect everyone.

# HTTP Requests/Responses

All HTTP Requests/Responses have some useful characteristics.



### HTTP Request Methods

There are some types of requests:
* CONNECT - used to create a tunnel between client and server
* GET - retrieves data
* POST - sends data to server
* PUT - creates new resource or replaces existing resource using complete representation
* PATCH - updates resource specifically
* DELETE - deletes resource on server side
* OPTIONS - requests permitted communication options from server
* TRACE - sends a message to server; useful for debugging


#### HTTP Headers

HTTP headers are sent with requests (generated on your end) or sent with response (generated on server side). Headers contain meaningful information about your req/resp.

Using `curl` to send a `GET` request to www.google.com, we can pull an example header:
    
```
HTTP/1.1 200 OK
Content-Type: text/html; charset=ISO-8859-1
P3P: CP="This is not a P3P policy! See g.co/p3phelp for more info."
Date: Mon, 02 Nov 2020 20:03:23 GMT
Server: gws
X-XSS-Protection: 0
X-Frame-Options: SAMEORIGIN
Transfer-Encoding: chunked
Expires: Mon, 02 Nov 2020 20:03:23 GMT
Cache-Control: private
Set-Cookie: NID=204=bdDRPoX7DCjujpYXM; expires=Tue, 04-May-2021 20:03:23 GMT; path=/; domain=.google.com; HttpOnly
```

The `curl` command used:

```bash
curl --header www.google.com
```

### HTTP Response Status Codes

Status code informs you of the server's response in categorical manner. This is useful in allowing us to handle different responses from the server.

E.g. if we send a `POST` request, we will get back a response that might be 201/CREATED or 400/BAD_REQUEST. If we get 201, we can continue our code. If we get 400, we need to handle our error.

````python
from http import HTTPStatus

for status_code in HTTPStatus:
    print(int(status_code), status_code)
````

````python
# Example of a POST request

import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:83.0) Gecko/20100101 Firefox/83.0',
    'Content-Type': 'application/x-www-form-urlencoded',
}
data = {
  'custname': 's',
  'custtel': 'd',
  'custemail': 'd@sdf.com',
  'size': 'small',
  'topping': 'bacon',
  'delivery': '12:15',
  'comments': 'a'
}

response = requests.post('https://httpbin.org/post', headers=headers, data=data)
print(response.text)
````
