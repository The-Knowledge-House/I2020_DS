# Project Day

At this point you should have started three separate projects.

These projects are as follows:

### A data analysis project

Researching a topic and exploring related data 

involving (at minimum):
* git repo
* readme file explaining your project, your analysis and how to set up and run your code 
* `/datasets` folder containing all of your data files
* any python files needed for your work
* jupyter notebooks 
  - one tracking your progress
  - a final report you'd present

### A machine learning project:

Displaying understanding of machine learning on some level

involving (at minimum):
* git repo
* readme file explaining your project, your model and how to set up and run your code 
* `/datasets` folder containing all of your data files
* all your python files needed for your work 
* jupyter notebook tracking progress and results

### A data engineering project

Building a data dashboard that visualizes data in useful ways

involving (at minimum)
* a git repo
* a django application with bokeh (or alternative)
* `/datasets` folder containing all of your data files and `sqlite` db containing your data
* deployed application hosted somewhere.
