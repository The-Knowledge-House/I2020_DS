<h3><a style="padding-top: 5px" href="#" id="2">What is Data Science?</a></h3>

#### Data science is an inter-disciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge and insights from many structural and unstructured data.[1][2] Data science is related to data mining, machine learning and big data.

#### Data science is a "concept to unify statistics, data analysis and their related methods" in order to "understand and analyze actual phenomena" with data.[3] It uses techniques and theories drawn from many fields within the context of mathematics, statistics, computer science, domain knowledge and information science. Turing award winner Jim Gray imagined data science as a "fourth paradigm" of science (empirical, theoretical, computational and now data-driven) and asserted that "everything about science is changing because of the impact of information technology" and the data deluge.[4][5]

Essentially, data science is using different tools and methods available to make sense of the data in order to understand it. There are different tools, methods and even types of data but the pursuit is the same. How can we take what we know and make sense of it? 

Remember this? __"What do we know? What do we need? How do we get there?"__

Charles Kettering, the famed inventor and head of research for GM, famously said 
> # a problem well-stated is half-solved.

The first step in data science is asking questions. A data scientist takes problems from the world and turns them into questions that can be answered with the tools/methods available.

e.g. some problems businesses have:
* Who is most likely to buy our product?
* What are the chances of success if we take on this project?
* Which areas of the business should we devote resources?
* How can we automate this decision making?

As a data scientist, you take these problems and formalize them into a question you have techniques to deal with. Asking the right questions is how you begin your analysis. We like to think of data scientists as programmers or mathematicians but they're scientists. They ask questions, form hypotheses and then test them in hopes of explaining phenomena. Setting up these experiments is often formalized as well.

### Workflow

The process that data scientists take in their research is called the workflow. There are various different workflows you can choose. 

From [Data Science Solutions](https://www.amazon.com/Data-Science-Solutions-Startup-Workflow/dp/1520545312):

The competition solution workflow goes through seven stages described in the Data Science Solutions book.
1. Question or problem definition.
2. Acquire training and testing data.
3. Wrangle, prepare, cleanse the data.
4. Analyze, identify patterns, and explore the data.
5. Model, predict and solve the problem.
6. Visualize, report, and present the problem solving steps and final solution.
7. Supply or submit the results.

From 'Introducing Data Science, Big Data, Machine Learning and More Using Python'
https://www.programmer-books.com/wp-content/uploads/2018/07/Introducing-Data-Science-Big-Data-Machine-Learning-and-more-using-Python-tools-2016.pdf



A well known industry pattern for business data analysis is __`CRISP DM`__:   

__CR__oss __I__ndustry __S__tandard __P__rocess for __D__ata __M__ining (CRISP-DM) is a process model with six phases that naturally describes the data science life cycle. Its like a set of guardrails to help you plan, organize, and implement your data science project.
1. Business understanding  What does the business need?
2. Data understanding  What data do we have / need? Is it clean?
3. Data preparation  How do we organize the data for modeling?
4. Modeling  What modeling techniques should we apply?
5. Evaluation  Which model best meets the business objectives?
6. Deployment  How do stakeholders access the results?

[more here](https://www.datascience-pm.com/crisp-dm-2/)



It's important to notice that this pattern can loop infinitely. It is important to set out goals when doing any sort of data analysis. Especially with "deployment" there, usually your analysis is needed within a time frame. It's important that your goals are in line with what is doable with the resources and time you have.

### Descriptive Statistics - 

#### Measures of Central Tendency: Mean, Median and Mode

##### __Mean or Expected Value__

$$ E[X] = \bar{X} = \frac{1}{n} \sum_{i=1}^{n} x_i $$

Essentially what this equation is saying is that the mean is equal to the sum of the numbers from 1 to n divided by the number of elements.

##### __Median__

$$ 1, 2, 3, 4, 5, 6, 7 $$

median: `4`

$$ 1, 2, 3, 4, 4, 5, 6, 7 $$

median: `4` $$ median = \frac{4+4}{2} $$ 

The median is the midpoint of the set of numbers. If the set is even, the median is the average of the two midpoint values.

##### __Mode__

$$ 1, 2, 3, 4, 3, 3, 5, 6$$

mode: `3`

The mode is the value that appears the most in the set.

#### __Standard Deviation__

The mean is the average value taken by summing the values of a set and dividing by the number of values in the set.

The standard deviation is the measure of the amount of variation from that mean. A high standard deviation means that there were a high deviation from the mean.

The variance is the square root of the standard deviation. Informally, it measures how far numbers deviate from the mean in absolute value (a square root must be positive).

$$ variance =  \frac{\sum{(x_i-\bar{X})}^2}{n-1}$$

$$ s =  \sqrt{\frac{\sum{(x_i-\bar{X})}^2}{n-1}} $$
